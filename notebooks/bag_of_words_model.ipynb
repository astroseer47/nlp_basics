Text Vectorization

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
data = ["this is a sentence one and it has many words and these words are just some words", "vampire drank the blood", "the bus arrives everyday at the same time"]
```

```python
countvec = CountVectorizer()
```

```python
countvec_fit = countvec.fit_transform(data)
```

```python
bag_of_words = pd.DataFrame(countvec_fit.toarray(), columns = countvec.get_feature_names_out())
bag_of_words.head()
```

```python
tf_idf = TfidfVectorizer()
tf_idf_fit = tf_idf.fit_transform(data)
```

```python
tfidf_bag = pd.DataFrame(tf_idf_fit.toarray(), columns = tf_idf.get_feature_names_out())
tfidf_bag.head()
```
