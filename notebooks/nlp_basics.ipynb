```python
string = r"sara was able to help me find the items I needed quickly"
new_string = re.sub("sara", "sarah", string)
print(new_string)
```

```python
customer_reviews = ["sam was great help. sam's great", "cashier was rude, i think her name is %eleanore!", "amazing work sid", "sarah helped me!!", "lucy is great", "great service from sara when i needed"]
sarah_reviews = []
pattern_to_find = r"sarah?"
for string in customer_reviews:
    if re.search(pattern_to_find, string):
        sarah_reviews.append(string)

print(sarah_reviews)
```

```python
a_reviews = []
pattern_to_find = r"^a"
for string in customer_reviews:
    if re.search(pattern_to_find, string):
        a_reviews.append(string)

print(a_reviews)
```

```python
y_reviews = []
pattern_to_find = r"y$"
for string in customer_reviews:
    if re.search(pattern_to_find, string):
        y_reviews.append(string)

print(y_reviews)
```

```python
need_want_reviews = []
pattern_to_find = r"(need|want)ed"
for string in customer_reviews:
    if re.search(pattern_to_find, string):
        need_want_reviews.append(string)

print(need_want_reviews)
```

```python
no_punct_reviews = []
pattern_to_find = r"[^\w\s]"
for string in customer_reviews:
    no_punct_string = re.sub(pattern_to_find,"", string)
    no_punct_reviews.append(no_punct_string)

print(no_punct_reviews)
```

```python
string = "Hi here, hi again and hi there too"
print(re.search("hi", string))
```

N-grams


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significat', 'advancement', 'in', 'natural', 'language', 'processing' 'to', 'to', 'two', 'years', 'to', 'come', 'grow', 'set', 'only', 'vision', 'used', 'being', 'of', 'humans','domain', 'tasks', 'in', 'are', "the", "rise", "of"]
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())

print(unigrams[:10])
```

```python
unigrams[:10].sort_values().plot.barh(color='lightsalmon', width=9, figsize=(12,8))
plt.title("10 most frequently occuring unigrams")
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())

print(bigrams[:10])
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())

print(trigrams[:10])
```
